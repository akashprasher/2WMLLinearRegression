```
# Reading and Displaying images in different formats using different colour models.

import numpy as np
import pandas as pd
import cv2 as cv
import matplotlib.pyplot as plt
```

```
img = cv.imread('flower.jpg', -1)
plt.imshow(img)
```

```
img = cv.imread('flower.jpg', 1)
plt.imshow(img)
```

```
img = cv.imread('flower.jpg', 0)
plt.imshow(img)
```

```
img = cv.imread('flower.jpg', 1)
imag = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(imag)
```

```
img = cv.imread('flower.jpg', 1)
imag = cv.cvtColor(img,cv.COLOR_BGR2HSV)
plt.imshow(imag)
```

```
img = cv.imread('flower.jpg', 1)
imag1 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
imag2 = cv.cvtColor(imag1,cv.COLOR_RGB2GRAY)
plt.imshow(imag2)
```
