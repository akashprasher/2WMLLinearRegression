```
import cv2 as cv
import matplotlib.pyplot as plt
```

```
image1 = cv.imread('/content/drive/MyDrive/0-For_Colab/flower_cu.jpg', -1)
plt.imshow(image1)
```

```
image2 = cv.imread('/content/drive/MyDrive/0-For_Colab/flower_cu.jpg', 1)
plt.imshow(image2)
```

```
image3 = cv.imread('/content/drive/MyDrive/0-For_Colab/flower_cu.jpg', 0)
plt.imshow(image3)
```

```
image4 = cv.imread('/content/drive/MyDrive/0-For_Colab/flower_cu.jpg', 1)
img1 = cv.cvtColor(image4, cv.COLOR_BGR2RGB)
plt.imshow(img1)
```

```
image = cv.imread('/content/drive/MyDrive/0-For_Colab/flower_cu.jpg', 1)
img2 = cv.cvtColor(image,cv.COLOR_BGR2HSV)
plt.imshow(img2)
```

```
image5 = cv.imread('/content/drive/MyDrive/0-For_Colab/flower_cu.jpg', 1)
img3 = cv.cvtColor(image5, cv.COLOR_BGR2GRAY)
plt.imshow(img3)
```

```
images_horozontal = cv.hconcat((img1, img2))
plt.imshow(images_horozontal)
```

```
images_vertical = cv.vconcat((img1, img2))
plt.imshow(images_vertical)
```
